```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import sklearn.datasets
import sklearn.linear_model
import sklearn.model_selection
```

```python
boston = sklearn.datasets.load_boston()
```

```python
# 住宅情報表示
X = boston.data
df = pd.DataFrame(boston.data, columns = boston.feature_names)
df
```

```python
y = boston.target  # 住宅価格
y
```

```python
# train：訓練データ, test：テストデータ
# test_size=0.2 (８０％訓練データ、20％検証用テストデータ)
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2)
```

```python
lr = sklearn.linear_model.LinearRegression()
```

```python
lr.fit(X_train, y_train)  # 訓練データを入れて、学習させる
```

```python
lr.score(X_test, y_test)  # アルゴリズムの正当性がどれくらい正しいか？叩き出す
```

```python
predicted = lr.predict(X)  # predictで予測
```

```python
fig, ax = plt.subplots()
ax.scatter(y, predicted, edgecolors=(0, 0, 0))
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
# 黒点線：住宅データに基づいた予測したデータ
```
